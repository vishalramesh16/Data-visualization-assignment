```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/visha/OneDrive/Documents/bigmart.csv")
df.head()
```

```python
df.info()
```

```python
plt.hist(df['Item_Weight'].dropna(), bins=20)
plt.xlabel('Item Weight')
plt.ylabel('Frequency')
plt.title('Histogram of Item Weight')
plt.show()
```

```python
sns.countplot(x='Item_Type', data=df)
plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Count')
plt.title('Countplot of Item Type')
plt.show()
```

```python
sns.countplot(x='Outlet_Type', data=df)
plt.xlabel('Outlet Type')
plt.ylabel('Count')
plt.title('Barplot of Outlet Type')
plt.show()
```

```python
outlet_size_counts = df['Outlet_Size'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(outlet_size_counts, labels=outlet_size_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Outlet Size')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
